```python
import pandas as pd
import numpy as np
```

```python
hoja1 = pd.read_excel('SIN CORREO V2 JULIO 2023\ReporteGenerado24072023.xlsx',sheet_name = 'HOJA 1')
hoja2 = pd.read_excel('SIN CORREO V2 JULIO 2023\ReporteGenerado24072023.xlsx',sheet_name = 'HOJA 2')
hoja3 = pd.read_excel('SIN CORREO V2 JULIO 2023\ReporteGenerado24072023.xlsx',sheet_name = 'HOJA 3')
hoja4 = pd.read_excel('SIN CORREO V2 JULIO 2023\ReporteGenerado24072023.xlsx',sheet_name = 'HOJA 4')
hoja5 = pd.read_excel('SIN CORREO V2 JULIO 2023\ReporteGenerado24072023.xlsx',sheet_name = 'HOJA 5')

consolidado = pd.concat([hoja1, hoja2, hoja3, hoja4, hoja5])
```

```python
declarantes = pd.read_excel('SIN CORREO V2 JULIO 2023\DEF_884565 DECLARANTES RENTA.xlsx')
```

```python
rut_sin_declarantes = pd.DataFrame(consolidado.loc[np.invert(consolidado['NIT'].isin(declarantes['NIT']))])
```

```python
rut_sin_declarantes
```

```python
cod_seccional = rut_sin_declarantes.COD_SECC.unique()
cod_seccional =cod_seccional.tolist()
print(cod_seccional)
```

```python
writer = pd.ExcelWriter('consolida2_rut.xlsx', engine ='xlsxwriter')
for i in rut_sin_declarantes.groupby('COD_SECC'):
    if i[0]> 32:
     i[1].to_excel(writer, sheet_name= f'Seccional_{i[0]}', index= False)
    # df_seccional=rut_sin_declarantes[rut_sin_declarantes['COD_SECC']==32]
    # subconjunto.to_excel(writer, sheet_name= f'Hoja {i+1}', index= False)
    # # df_seccional.to_csv(f"{i}.csv",index=False, encoding="latin-1")

```

```python
for i in cod_seccional:
    df_seccional=rut_sin_declarantes[rut_sin_declarantes['COD_SECC']==32]
    df_seccional.to_csv(f"32.csv",index=False, encoding="latin-1")
display(df_seccional)
```

```python
num_divisiones = len(df_seccional)//500000+1
writer = pd.ExcelWriter('Seccional_32.xlsx', engine ='xlsxwriter')
for i in range(num_divisiones):
    inicio = i*500000
    fin = (i+1)*500000
    subconjunto = df_seccional[inicio:fin]
    subconjunto.to_excel(writer, sheet_name= f'Hoja {i+1}', index= False)
```
